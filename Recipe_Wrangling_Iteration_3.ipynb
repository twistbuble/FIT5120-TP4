### Measurement Regex

```python
fahrenheit_re = re.compile("(°F)")
fahrenheit_va = re.compile(r"(\d+)°F")

# ounces_re = re.compile("(\d*\s*\d+.\d*\s*oz.|\d*\s*\d+.\d*\s*ounce[s]*)")
# pounds_re = re.compile("(\d+.pound[s]*)")
```

### Converting Fahrenheit to Celcius

```python
def numrepl(matchobj):
    return "{}°F".format( int(round((int(matchobj.group(1))-32)*5/9,0)) )

def degree_change(df):
    for index, series in df.iterrows():
        result = fahrenheit_va.sub( numrepl, df.loc[index, "Ingredients"])
        df.loc[index, "Ingredients"] = result
        degree = re.sub(fahrenheit_re, "°C", df.loc[index, "Ingredients"])
        df.loc[index, "Ingredients"] = degree
        result_in = fahrenheit_va.sub( numrepl, df.loc[index, "Instructions"])
        df.loc[index, "Instructions"] = result_in
        degree_in = re.sub(fahrenheit_re, "°C", df.loc[index, "Instructions"])
        df.loc[index, "Instructions"] = degree_in
    return df

degree_change(sampled_recipes)
```
